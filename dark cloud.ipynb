Perhitungan bintang menjadi dasar dalam menentukan jumlah bintang yang tampak dalam rentang magnitudo tertentu pada sebuah daerah yang terindikasi terdapat materi antar bintang.

Berikut adalah hasil perhitungan bintang dari Horsehead nebula pada bagian dark cloud dan field star, dengan mengambil beberapa daerah dalam radius 1 derajat. 

![title](all.jpg)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cloud = pd.read_csv('cloud')
all_star = pd.read_csv('all')
field1 = pd.read_csv('field1')
field2 = pd.read_csv('field2')
```

```python
cloud['average']=(cloud['LOW']+cloud['HIGH'])/2
all_star['average']=(all_star['LOW']+all_star['HIGH'])/2
field1['average']=(field1['LOW']+field1['HIGH'])/2
field2['average']=(field2['LOW']+field2['HIGH'])/2
```

```python
from numpy.polynomial.polynomial import polyfit
b1,m1 = polyfit(field2['average'],field2['field2_COUNT']/len(field2['field2_COUNT']),1)
b2,m2 = polyfit(cloud['average'],cloud['cloud1_COUNT']/len(cloud['cloud1_COUNT']),1)
```

```python
plt.scatter(field2['average'], field2['field2_COUNT']/len(field2['field2_COUNT']),marker = '.', label = 'field')
plt.plot(field2['average'], b1+m1*field2['average'], '-', label = 'field regression')
plt.scatter(cloud['average'], cloud['cloud1_COUNT']/len(cloud['cloud1_COUNT']),marker = '.', label = 'cloud')
plt.plot(cloud['average'], b2+m2*cloud['average'], '-',label = 'cloud regression')
plt.title('Wolf Diagram')
plt.legend()
plt.xlabel('mag')
plt.ylabel('N normalized')
```
